```python
#yeild example to print even numbers from a list

def print_even(test_list) :
    for i in test_list:
        if i % 2 == 0:
            yield i

test_list = [1, 4, 5, 6, 7]

print ("The original list is : " + str(test_list))

# printing even numbers
print ("The even numbers in list are : ", end = " ")
for j in print_even(test_list):
    print (j, end = " ")
```
